```python
import pandas as pd

num = 16
results = pd.read_csv(f"train16/results.csv")
```

```python
results.head(20)
```

```python
import matplotlib
matplotlib.use('TKAgg')
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(results["epoch"], results["train/box_loss"], color="k", label="training")
plt.plot(results["epoch"], results["val/box_loss"], color="red", label="validation")
plt.xlabel("Epochs")
plt.ylabel("Loss (Box)")
plt.legend()
plt.savefig(f"Loss-Box-{num}")
plt.show()
```

```python

plt.plot(results["epoch"], results["train/cls_loss"], color="k", label="training")
plt.plot(results["epoch"], results["val/cls_loss"], color="red", label="validation")
plt.xlabel("Epochs")
plt.ylabel("Loss (Class)")
plt.legend()
plt.savefig(f"Loss-Class-{num}")
plt.show()
```

```python
plt.plot(results["epoch"], results["train/dfl_loss"], color="k", label="training")
plt.plot(results["epoch"], results["val/dfl_loss"], color="red", label="validation")
plt.xlabel("Epochs")
plt.ylabel("Loss (dfl)")
plt.legend()
plt.savefig(f"Loss-dfl-{num}")
plt.show()
```

```python
plt.plot(results["epoch"], results["metrics/recall(B)"], color="k")
plt.xlabel("Epochs")
plt.ylabel("Recall")
plt.savefig(f"Recall-{num}")
plt.show()
```

```python
plt.plot(results["epoch"], results["metrics/precision(B)"], color="k")
plt.xlabel("Epochs")
plt.ylabel("Precision")
plt.savefig(f"Precision-{num}")
plt.show()
```
